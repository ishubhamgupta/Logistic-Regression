# Data Cleaning

```python
sns.boxplot(x='Pclass', y='Age', data=train)
```

```python
plt.figure(figsize=(12,9))
sns.boxplot(x='Pclass', y='Age', data=train)
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else: 
            return 24
    else:
        return Age
# By looking we are assuming the ages
```

```python
train['Age'] = train[['Age', 'Pclass']].apply(impute_age, axis=1)
```

```python
sns.heatmap(train.isnull(), yticklabels= False, cbar= False)
```

```python
sns.heatmap(train.isnull(), yticklabels= False, cbar= False, cmap = 'viridis')
```

```python
train.drop('Cabin', axis =1, inplace=True)
```

```python
sns.heatmap(train.isnull(), yticklabels= False, cbar= False, cmap = 'viridis')
```

```python
train.dropna(inplace=True)
```

```python
sns.heatmap(train.isnull(), yticklabels= False, cbar= False, cmap = 'viridis')
```

```python
# Categorical Features

pd.get_dummies(train['Sex'])

```

**One column is the perfect predictor of the other column, meaning if a ML algorithm gets fed in both the columns the ML algorithm
 if its 0 at 'female' I can immediately predict its 'male'.  This issue is known as "Multi Colinearity", and it will basically mess 
     up the algorithm**

```python

pd.get_dummies(train['Sex'],drop_first=True)
```

```python
sex = pd.get_dummies(train['Sex'],drop_first=True)
```

```python
sex.head()
```

```python
embark = pd.get_dummies(train['Embarked'], drop_first=True)
```

```python
embark.head()

# Although we have droped the 'C' columns but we have to understand all columns are not perfect predictors of each other
```

```python
 train = pd.concat([train,sex,embark], axis=1)
```

```python
train.head()
# We have encoded the 'SEX' to Male
# We have encoded the 'Embarked' to Q & S
```

```python
train.drop(['Sex', 'Embarked', 'Name','Ticket'], axis=1, inplace=True)
```

```python
train.head()
```

```python
train.tail()
```

```python
train.drop('PassengerId', axis=1, inplace=True)
```

```python
train.head()
```

**We can do the dummies for 'Pclass' as well as it only has three value**

```python
X = train.drop('Survived', axis=1)
y = train['Survived']
# Y is the column that we are trying to predict
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
predictions =logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
```
